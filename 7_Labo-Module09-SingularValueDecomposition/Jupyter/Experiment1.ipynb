```python
import numpy as np
from numpy import genfromtxt

# Load signal
data = genfromtxt('ECG.csv', delimiter=',')

# Perform SVD
U1, S1, V1 = np.linalg.svd(data[:, 0:6], full_matrices=False)
print(S1)

# Perform SVD
U2, S2, V2 = np.linalg.svd(data[:, 6:12], full_matrices=False)
print(S2)
```

```python
# Add here the code to compute the effective rank
def effective_rank(S, threshold=0.98):
    # compute the t of after truncating each singular value
    energy = np.cumsum(S**2) / np.sum(S**2)
    # find the rank where the cumulative energy exceeds the threshold
    return np.where(energy >= threshold)[0][0] + 1

rank1 = effective_rank(S1)
print("Effective rank (first 6 columns):", rank1)

rank2 = effective_rank(S2)
print("Effective rank (last 6 columns):", rank2)
```
